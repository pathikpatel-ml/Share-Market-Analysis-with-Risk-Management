```python
import matplotlib.pyplot as plt 
from datetime import date 
from nsepy import get_history 
for i in a:
         
        stock_opt = get_history(symbol=i, 
                        start=date(2018,6,1), 
                        end=date(2020,12,15),)
        stock_opt.Close.plot()
        plt.xlabel('Date')
        plt.ylabel('Prev Close')
        plt.title(i + '  ' + 'stock graph')
        plt.grid(True)
        plt.ylim(10,1500)
        plt.show() 
        stock_opt.head(10)
```

```python
stock_opt.tail.avg()
```

```python
df = stock_opt 
df.VWAP
```

```python
df['simple_return']= (df['VWAP'] / df['VWAP'].shift(1)) - 1
print (df['simple_return'])
```

```python
df['simple_return'].plot(figsize=(16,5))
plt.show()
```

```python
avg_return_d= df['simple_return'].mean()
avg_return_d
```

```python
avg_return_a= df['simple_return'].mean()*250
avg_return_a
```
